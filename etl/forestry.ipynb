```python
all_data = []

for f in os.listdir(source_dir):
    if 'csv' in f:
        if f == '9. ANNUAL DATA.csv':
            df = pd.read_csv(os.path.join(source_dir, f), 
                             encoding='iso-8859-1', 
                             sep=';', decimal=',', 
                             skipinitialspace=True,
                             na_values=['n.a.', 'n/a']
                            )
        else:
            df = pd.read_csv(os.path.join(source_dir, f),
                             encoding='iso-8859-1', 
                             skipinitialspace=True,
                             na_values=['n.a.', 'n/a'],
                             thousands=','
                            )

        df = df.loc[:, 'Country':]
        
        all_data.append(df)
```
